```python
max_town = 1541

df_info = pd.DataFrame(columns=['Phone number', 'Fax', 'Email', 'Website', 'Address', 'locality', 'administrative_area_level_2','administrative_area_level_1', 'country'])


with tqdm_notebook(range(max_town+1)) as pbar:
    for ii in range(1,max_town+1):
        url = "http://interieur.gov.dz/index.php/fr/collectivit%C3%A9s-territoriales/rechercher-une-collectivite-locale/annuairecommune/{}.html".format(ii)
        response = requests.get(url)  
        soup = BeautifulSoup(response.text, "html.parser")
        
        
        commune=soup.find_all("h2")[0].get_text()
        commune=commune.replace("Commune de ",'').strip()
        commune=commune.title()
        tel=soup.find_all("td")[14].get_text()
        fax=soup.find_all("td")[15].get_text()
        email=soup.find_all("td")[16].get_text().lower()
        website=soup.find_all("td")[17].get_text()
        address=soup.find_all("td")[18].get_text()
        address=address.strip()
        address=address.rstrip()
        daira=soup.find_all("td")[19].get_text().title()
        daira=daira.strip()
        wilaya=soup.find_all("td")[20].get_text().title()
        wilaya=wilaya.strip()

        
        df_info=df_info.append(
        {'Phone number': tel, 'Fax': fax, 'Email': email, 'Website':website, 'Address':address, 'locality': commune, 'administrative_area_level_2':daira, 'administrative_area_level_1':wilaya, 'country': 'Algeria'}, 
        ignore_index=True)
        
        pbar.update()
```

```python
df_info.to_csv("list_cities.csv", index=False)
```
